```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# Assuming X and y are already defined
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.2, random_state=42)

# Initialize the model with warm start so trees are added incrementally
clf = RandomForestClassifier(n_estimators=1, warm_start=True, random_state=42, min_samples_split=5)

train_accuracies = []
test_accuracies = []
num_trees = 10  # Number of trees to track performance

for i in range(1, 50):
    clf.n_estimators = i  # Incrementally add trees
    clf.fit(X_train, y_train)  # Train with updated trees
    
    # Make predictions
    y_train_preds = clf.predict(X_train)
    y_test_preds = clf.predict(X_test)

    # Calculate accuracy
    train_acc = accuracy_score(y_train, y_train_preds)
    test_acc = accuracy_score(y_test, y_test_preds)

    # Display accuracy at each step
    print(f"Tree {i}: Train Acc = {train_acc:.4f}, Test Acc = {test_acc:.4f}")
```

```python
from sklearn.model_selection import cross_val_score

cv_scores = cross_val_score(clf, X, y, cv=5)  # 5-fold cross-validation
print("Cross-validation accuracy:", cv_scores.mean())
```
